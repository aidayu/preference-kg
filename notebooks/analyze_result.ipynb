```python
result_path = "results/output_from_gpt4o-mini_SU_20251223_022824.json"
data_path = "/home/y-aida/Programs/preference-kg/data/interim/dailydialog_annotated_SU_20251223_022000_aida.json"
result_dir = "results/"
```

```python
import json
import sys
import os

# preference_kgモジュールをインポートできるようにパスを追加
notebook_dir = os.getcwd()  # notebookの実行ディレクトリ
project_root = os.path.dirname(notebook_dir)  # プロジェクトルート
if project_root not in sys.path:
    sys.path.insert(0, project_root)

from preference_kg.evaluate_mvp import run_extraction, evaluate_predictions, print_metrics, save_metrics, compare_annotations_detailed
```

```python
with open(result_path, "r", encoding="utf-8") as f:
    llm_outputs = json.load(f)
```

```python
annotation_counts = 0
liking_counts = 0
liking_aesthetic_counts = 0
liking_stimulation_counts = 0
liking_identification_counts = 0
liking_general_counts = 0
wanting_counts = 0
wanting_interest_counts = 0
wanting_goal_counts = 0
need_counts = 0
need_functional_counts = 0
need_social_counts = 0
need_personal_counts = 0

for entry in llm_outputs:
    for pref in entry["predictions"]["preferences"]:
        annotation_counts += 1
        axis = pref["axis"]
        sub_axis = pref["sub_axis"]
        if axis == "liking":
            liking_counts += 1
            if sub_axis == "aesthetic/sensory":
                liking_aesthetic_counts += 1
            elif sub_axis == "stimulation":
                liking_stimulation_counts += 1
            elif sub_axis == "identification":
                liking_identification_counts += 1
            else:
                liking_general_counts += 1
        elif axis == "wanting":
            wanting_counts += 1
            if sub_axis == "interest":
                wanting_interest_counts += 1
            elif sub_axis == "goal":
                wanting_goal_counts += 1
        elif axis == "need":
            need_counts += 1
            if sub_axis == "functional":
                need_functional_counts += 1
            elif sub_axis == "social":
                need_social_counts += 1
            elif sub_axis == "personal":
                need_personal_counts += 1
                
print(f"Total dialogue entries processed: {len(llm_outputs)} ")
print(f"Total annotations processed: {annotation_counts} ")
print(f"Total 'liking' preferences: {liking_counts}")
print(f"  Aesthetic: {liking_aesthetic_counts}")
print(f"  Stimulation: {liking_stimulation_counts}")
print(f"  Identification: {liking_identification_counts}")
print(f"  General: {liking_general_counts}")

print(f"Total 'wanting' preferences: {wanting_counts}")
print(f"  Interest: {wanting_interest_counts}")
print(f"  Goal: {wanting_goal_counts}")

print(f"Total 'need' preferences: {need_counts}")
print(f"  Functional: {need_functional_counts}")
print(f"  Social: {need_social_counts}")
print(f"  Personal: {need_personal_counts}")
```

```python
# データセットを読み込む
with open(data_path, "r", encoding="utf-8") as f:
    dataset = json.load(f)

# 評価を実行
metrics = evaluate_predictions(dataset, llm_outputs)
print_metrics(metrics)
save_metrics(metrics, extraction_type="zero_shot", result_dir=result_dir)
```

```python
# すべての対話を詳細比較
results = compare_annotations_detailed(dataset, llm_outputs)
```
